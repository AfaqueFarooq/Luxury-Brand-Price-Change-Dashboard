```python
import pandas as pd

w1 = pd.read_excel('LY Case Study Quantitative_11.11.24.xlsx',sheet_name='W-1')

w = pd.read_excel('LY Case Study Quantitative_11.11.24.xlsx',sheet_name='W')
```

```python
w1.columns
```

```python
w.columns
```

```python
df = pd.merge(w1,w, on='SKU' , suffixes=('_w1','_w'))
```

```python
df.columns
```

```python
df.head()
```

```python
countries = [
    'France', 'United Kingdom', 'USA (tax free)', 'Japan', 'Hong Kong', 'China', 
    'Korea', 'Korea DF', 'Hainan DF', 'United Arab Emirates', 'Australia', 'Canada', 
    'Macao', 'Mexico', 'Saudi Arabia', 'Singapore', 'Thailand', 'Taiwan'
]

for country in countries:
    col_w = f'{country}_w'
    col_w1 = f'{country}_w1'

    df[f'{country} Price Diff'] = df[col_w] - df[col_w1]

    df[f'{country} Pct-Change'] = ((df[col_w]-df[col_w1])/df[col_w1])*100
    df[f'{country} Pct-Change'] = df[f'{country} Pct-Change'].fillna(0)

    
```
